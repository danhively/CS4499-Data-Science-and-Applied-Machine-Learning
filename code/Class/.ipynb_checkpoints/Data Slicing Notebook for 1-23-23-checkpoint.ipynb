# Data Slicing with NumPy

```python
#import packages 
import numpy as np
```

```python
#Create 1D ndarray 
#We will the arrange method to create our data 
myData = np.arange(0, 10000, 99)
print(myData)
```

```python
myData.size
```

```python
#Basic slice 
#list the first five entries in our ndarray 
mySlice= myData[0:5]
mySlice
#Note this will not include the final entry 
```

```python
#If we leave the end point black it will default to 0 
mySlice_2 = myData[:5]
mySlice_2
```

```python
#If we the end point empty it will default to n 
mySlice_2= myData[97:]
mySlice_2
```

```python
#Let's do an example with the step 
mySlice_2=myData[::10]
mySlice_2
```

# Slicing with NumPy 

Now lets mess around with some real data that is hosted online.

```python
#Here we are just importing some data to work with and setting it up as a ndarray (we will go over the ins and outs a little late ). There are 573 data points 
import pandas as pd
salariesdf = pd.read_csv('https://gitlab.com/CEADS/DrKerby/python/raw/master/NBA_season1718_salary.csv')
salaries = salariesdf['season17_18'].values
salaries
```

```python
type(salaries)
```

```python
#We can check the number of entries using len() just like we would with a Python List 
len(salaries)
```

```python
#Lets sort our data 
salaries.sort(kind="mergesort")
salaries 
```

```python
#If we place a negative number as our index, numpy will just revert back to the end of the array 
print(salaries[-1])
print(salaries[-2])
```

```python
#Let's print out the first 10 salaries from our datset 
print(salaries[0:10])
```

```python
print(salaries[-10:])
```

```python
#If our array isn't sorted and we don't want to sort we can find the max using .max()
#NumPy Method 
salaries.max()
#Python function 
max(salaries)
```

```python
#We can do min the same way 
#NumPy Method 
salaries.min()
#Python function 
min(salaries)
```

```python
#We can do basic math functions with numpy 
salaries.sum()
```

```python
#We use slices with most methods 
#we can add up the last 10 entries in the array 
salaries[-10:].sum()
```

```python
#We can peform basic math on the entire array 
#Divide all entries our by one million 
salaries_in_million=salaries/1000000
print(salaries_in_million)
```

```python
#We can calculate basic stats as well 
print(salaries.mean())
print(np.median(salaries))
```

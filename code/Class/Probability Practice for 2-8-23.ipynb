# Dungeons & Dragons Example 

In D&D each player plays fantasy characters. 

A character performs actions by rolling a dice a 20-side, add a "modifier" number to the roll based on a number of other factors. Then we compare the result of the roll to some threshold. 

For example, our player, Daniel, is playing as a ranger. He is trying sneak through a camp of monsters. He is going to roll 20-sided die and add modifier of -5, in order to be sucessful he needs a role higher than 11. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#Step 1: Simulate our 20 sided dice roll. 
possible_rolls = np.array(np.arange(20)) +1 
np.random.choice(possible_rolls)
#np.random.choice(20)
```

```python
#Step 2: Apply our modifier 
modifer = -5 
final_roll = np.random.choice(20) + 1 + modifer
final_roll
```

```python
#Step 3: Check for sucess 
print(f'Final roll: {final_roll}; Succeeded: {final_roll>11}')
```

```python
#Let's run this 7 times and count the number of successs and failures 
succeed = 0 
fail = 0 
modifier =-5
for i in range(10000):
  final_roll = np.random.choice(20) + 1 + modifer
  #print(final_roll)
  if (final_roll >=11):
    succeed=succeed+1
  else:
    fail = fail + 1 
print('Total Number of passed rolls: ' + str(succeed))
print('Total Number of failed rolls:' + str(fail))
```

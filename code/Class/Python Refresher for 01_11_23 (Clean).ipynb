# Loops in Python 

Loops are another common operationin Python. Like conditional statements, Python makes use use of colons and indentations. In this class we will likely only use for loops, but we will go over while loops just in case. 

```python
#While loop example 
increment =0 
while (increment <10):
  print(increment)
  increment = increment+1 #Note make sure your loop always works toward the end condition
  #we can add a break statement to get out of the loop early 
  if(increment==5):
    break
```

```python
#For loop Example
#At first for loops can be a bit jarring if you are more familiar with Java, C++, etc. 
#With Python for loops we usualy make use of the in key word to specify the increments. 
#We can use the range function to specify our start, end and increment.
for i in range (4,10,2):#Note the default starts at 0 and has an increment of 1
  print(i)
```

#Classes in Python

In this class you will not have to make your own classes. However, we will be using many objects and methods throughout the class. Also in most neural network packages make use of class based models, so it is good to have this background. Most classes you will see will make use of an init method (a constructor). 

```python
class Person: 
  def __init__(self, name, age): #All classes should have this type of method 
  #self is just a variable we use to refer to our object within the class 
    self.name = name 
    self.age =age 
  def getName(self):
    return self.name 
  def getAge(self):
    return self.age 
  def summary(self):
    print("Hello my name is: " + self.getName() + ". " + "My age is: "
     + str(self.age))
```

```python
Pepo = Person("Pepo", 34)
Pepo.summary()
```

```python
#Get packages 
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

Now we need to clean the data. 

```python
pay_i=['PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
pay_1 = ['Pay_' +str(i) for i in range(1,7)]
for pay in pay_i:
  credit.loc[credit[pay] <=0, pay]=0
credit.head()
```

```python
#The ID # won't be useful so let's drop it 
credit.drop('ID', axis=1, inplace=True)
credit.head()
```

```python
credit['EDUCATION'].value_counts()
```

```python
credit.loc[credit['EDUCATION'] > 4, 'EDUCATION']= 4
credit.loc[credit['EDUCATION'] == 0, 'EDUCATION']=4
credit['EDUCATION'].value_counts()
```

```python
#Create a feature for how many customers have been late with payment over the last 6 months 
credit['MONTHS_DELAYED'] = credit[pay_i].sum(axis=1)
credit.head()
```

```python
#Categorical Features 
credit = pd.get_dummies(credit, columns=['EDUCATION', 'MARRIAGE'])
credit.head()
```

```python
#Fix SEX Feature 
credit.loc[credit['SEX'] == 1, 'SEX']= 0
credit.loc[credit['SEX'] == 2, 'SEX']= 1
credit.rename(columns={'SEX': 'FEMALE'}, inplace= True)
credit.head()
```

```python
#Find Probabilities Start with P(default)
P_def = credit['default payment next month'].mean()
P_def
```

```python
#P(male)
credit['FEMALE'].mean()
P_male = 1 - credit['FEMALE'].mean()
P_male
```

```python
#P(male | default)
defaults = credit[credit['default payment next month']==1]
P_male_def = 1 -defaults['FEMALE'].mean()
P_male_def
```

```python
#Now use Bayes to P(default | male)
P_def_male = P_male_def * P_def/P_male
P_def_male
```

```python
from sklearn.model_selection import train_test_split
X = credit.drop('default payment next month', axis=1)
y = credit['default payment next month']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
print('Training Acc', gnb.score(X_train, y_train))
print('Testing Acc', gnb.score(X_test, y_test))
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(X_train, y_train)
print('Training Acc', bnb.score(X_train, y_train))
print('Testing Acc', bnb.score(X_test, y_test))
```

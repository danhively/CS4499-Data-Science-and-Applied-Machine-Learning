# Data Preprocessing 

# Ordinal Data 
In these situations the magnitude of our data is important. So for it to work with Sklearn we need to convert it to numerical form. 

```python
#import packages 
import numpy as np 
import pandas as pd 
```

```python
#Create a DataFrame with some data 
df = pd.DataFrame({"Score": ['low', 'low', 'medium', 'medium', 'high', 'low', 'medium', 'high', 'low']})
df
```

We could just use a for loop to apply, but .replace() should be faster. 

```python
scale = {'low': 1, 'medium':2, 'high':3}
```

```python
df['Score Scale']= df['Score'].replace(scale)
df
```

```python
df.drop(['Score'], axis=1, inplace=True)
df
```

# Dummy Variables 
Pandas has a handy method called .get_dummies() that will convert your data into dummy varaible form. 

```python
#Let's bring in our Salary Data 
income = pd.read_csv('https://raw.githubusercontent.com/LGKerby/Python/master/en-2016-pssd-compendium.csv')
income
```

```python
income['Sector'].nunique()
```

```python
income_dummies = (pd.get_dummies(income['Sector']))
income_dummies
```

```python
income_dummies = (pd.get_dummies(income['Sector'])).iloc[:,1:]
income_dummies
```

```python
#Adding the dummy variables to the income data 
modified_income= pd.concat([income, income_dummies], axis=1)
modified_income
```

```python
#Get rid of sector column
modified_income.drop(['Sector'], axis=1, inplace=True )
modified_income
```

# Feature Scaling 
Feature scaling is one of the most important steps in prepearing data for training with machine learning models. In practice you will likely have to try out a number of scaling techniques. Let's go over a some common examples. 

```python
#Create some data to scale 
data = pd.DataFrame([[100,0.001], [50,0.005], [8,0.05], [88, 0.07], [4, 0.1]], columns=['X','Y'])
data
```

```python
#Define our Standard Scaler instance 
from sklearn.preprocessing import StandardScaler
Standard_scaler = StandardScaler()
#Scale the data 
Scaled_Data = Standard_scaler.fit_transform(data)
Scaled_Data = pd.DataFrame(Scaled_Data, columns=['', 'Y'])
Scaled_Data
```

```python
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.scatter(data['X'], data['Y'])
ax2.scatter(Scaled_Data['X'], Scaled_Data['Y'])
```

```python
data.std()
```

```python
#Robust Scaler Example 
#Distribution with lower outliers
new_data = pd.DataFrame({'x1': np.concatenate([np.random.normal(20,2,1000),np.random.normal(1,2,25)]),
                        #Distrubtion with higher outliers 
                        'x2': np.concatenate([np.random.normal(30,2,1000), np.random.normal(50,2,25)])}

                        )
 
```

```python
new_data
```

```python
#Robust Scaler 
from sklearn.preprocessing import RobustScaler
Robust_scaler = RobustScaler()
#Scale the data 
Robust_Scaled_Data = Robust_scaler.fit_transform(new_data)
Robust_Scaled_Data = pd.DataFrame(Robust_Scaled_Data, columns=['x1', 'x2']
)
Robust_Scaled_Data
```

```python
#Scale with Standard Scaler 
Scaled_Data = Standard_scaler.fit_transform(new_data)
Scaled_Data = pd.DataFrame(Scaled_Data, columns=['x1', 'x2'])
```

```python
import seaborn as sns 
fig, (ax1, ax2, ax3) = plt.subplots(ncols = 3, figsize =(20, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(new_data['x1'], ax = ax1, color ='r')
sns.kdeplot(new_data['x2'], ax = ax1, color ='b')
ax2.set_title('After Robust Scaling')
  
sns.kdeplot(Robust_Scaled_Data['x1'], ax = ax2, color ='red')
sns.kdeplot(Robust_Scaled_Data['x2'], ax = ax2, color ='blue')
ax3.set_title('After Standard Scaling')
sns.kdeplot(Scaled_Data['x1'], ax = ax3, color ='black')
sns.kdeplot(Scaled_Data['x2'], ax = ax3, color ='g')
```

```python
#Let's expand our example with the min-max scaler 
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler()
mm_data = mm_scaler.fit_transform(new_data)
mm_data = pd.DataFrame(mm_data, columns=['x1', 'x2'])
```

```python
fig, (ax1, ax2, ax3,ax4) = plt.subplots(ncols = 4, figsize =(20, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(new_data['x1'], ax = ax1, color ='r')
sns.kdeplot(new_data['x2'], ax = ax1, color ='b')
ax2.set_title('After Robust Scaling')
  
sns.kdeplot(Robust_Scaled_Data['x1'], ax = ax2, color ='red')
sns.kdeplot(Robust_Scaled_Data['x2'], ax = ax2, color ='blue')
ax3.set_title('After Standard Scaling')
sns.kdeplot(Scaled_Data['x1'], ax = ax3, color ='black')
sns.kdeplot(Scaled_Data['x2'], ax = ax3, color ='g')
ax4.set_title('After Min-Max Scaling')
sns.kdeplot(mm_data['x1'], ax =ax4, color ='black')
sns.kdeplot(mm_data['x2'], ax =ax4, color ='green')
```

```python
#Normalization Example 
X= [[4,1,2,2], [1,3,-9,3],[5,7,5,-1]]
```

```python
from sklearn.preprocessing import Normalizer 
#L2 Normalization 
norm_data = Normalizer().fit_transform(X)
norm_data
```

```python
#L1 Normalize 
norm_data_l1 = Normalizer(norm='l1').fit_transform(X) 
norm_data_l1
```

```python
#Max Absolute Scaling 
from sklearn.preprocessing import MaxAbsScaler
max_abs_data = MaxAbsScaler().fit_transform(X)
print(X)
max_abs_data
```

```python
#With Max ABS scaler we scaler by column (feature)
#If we want to scale by row we use the Max Normalizer 
norm_data_max = Normalizer(norm='max').fit_transform(X)
print(X)
norm_data_max
```

```python
#Binary Scaling 
from sklearn.preprocessing import Binarizer
#Default is threshold is 0
binary_data = Binarizer().fit_transform(X)
binary_data
```

```python
binary_data_2 = Binarizer(threshold=3).fit_transform(X)
binary_data_2
```
